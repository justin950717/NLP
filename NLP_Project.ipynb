```python
import re
import string

def remove_digits(text):
    return re.sub("\d+", "", text)

def remove_punctuation(text):
    return re.sub(f'[{string.punctuation}]', ' ', text)

def lowercase(text):
    return text.lower()
```

```python
df = df.rename(columns = {0: "texts"})
```

```python
df.head()
```

```python
url_pattern = '(http[s]?://)?[a-zA-Z0-9]+([-.][a-zA-Z0-9]+)*\.[a-zA-Z]{2,3}(/\S*)?'
df['contains_url'] = df['texts'].str.contains(url_pattern)
```

```python
df['contains_url'].value_counts()
```

```python
df['texts'] = df['texts'].str.replace(url_pattern, '', regex=True)
```

```python
url_pattern_2 = r'http\S+|www.\S+'

df['texts'] = df['texts'].str.replace(url_pattern_2, '', regex=True)
```

```python
df['texts'] = df['texts'].apply(remove_digits)
```

```python
df['texts'] = df['texts'].apply(remove_punctuation)
```

```python
df['texts'] = df['texts'].apply(lowercase)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
```

```python
stop_words
```

```python
df['stop_words_removal'] = df['texts'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stop_words]))
```

```python
df.head(20)
```

```python
df.tail(50)
```

```python
pattern = re.compile(r'^n$')

for row in df['stop_words_removal']:
    if pattern.match(row):
        print(row)
```

```python
mask = df['stop_words_removal'].apply(lambda x: True if pattern.match(x) else False)
```

```python
df = df.drop(df[mask].index)
```

```python
df.tail(50)
```

```python
df.reset_index(inplace = True, drop = True)
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

def lemmatize_text(text):
    return ' '.join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
df['lemmatized_text'] = df['stop_words_removal'].apply(lemmatize_text)
```
